```python
import nltk
from nltk.probability import FreqDist
from collections import Counter
nltk.download('punkt')
```

```python
def perform_eda(text):
    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Get frequency distribution
    fdist = FreqDist(tokens)

    # Display the most common tokens and their counts
    print("Most common tokens:", fdist.most_common(10))

    # Get the number of unique tokens
    print("Number of unique tokens:", len(set(tokens)))

    # Get the total number of tokens
    print("Total number of tokens:", len(tokens))

    # Get the lexical diversity
    print("Lexical diversity:", len(set(tokens)) / len(tokens))
```

```python
text = "Natural language processing helps computers communicate with humans in their own language and scales other language-related tasks."
perform_eda(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def create_wordcloud(text):
    # Create a WordCloud object
    wordcloud = WordCloud(width = 500, height = 500,
                background_color ='white',
                stopwords = None,
                min_font_size = 10).generate(text)

    # Plot the WordCloud image
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    plt.show()

text = "Natural language processing helps computers communicate with humans in their own language and scales other language-related tasks.  For example, NLP makes it possible for computers to read text, hear speech, interpret it, measure sentiment and determine which parts are important."
create_wordcloud(text)
```

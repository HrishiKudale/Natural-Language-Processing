```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```python
def pos_tagging(text):
    # Tokenize the text
    tokenized_text = nltk.word_tokenize(text)

    # Perform POS tagging
    pos_tags = nltk.pos_tag(tokenized_text)

    return pos_tags
```

```python
text = "It is a sunny weather today"
print(pos_tagging(text))
```

Using nltk stemming function

```python
import nltk
```

```python
text = 'working,worker,running,runner,taking,giving,writing'
```

```python
words = text.split(',')
print(words)
```

```python
stemmer = nltk.PorterStemmer()
```

```python
#Applying stemming on  words
[stemmer.stem(i) for i in words]
```
